## Definitionen
**Geodaten** (auch *spatial data*, *geospatial data* oder *raumbezogene Daten*) sind Daten, die mit einem Ort (auf der Erde) in Verbindung stehen. Sie enthalten typischerweise Koordinaten (z.B. Längen und Breitengrad) oder eine andere räumlich verordenbare Referenz. Zu letzteren zählen beispielsweise Postleitzahlen und IP-Adressen.

Wir unterscheiden dabei zwischen zwei verschiedenen **Koordinatensystemen**:
- Das **geographische Koordinatensystem** basiert auf Kugelkoordinaten (Länge und Breite in Grad).
- Das **projizierte Koordinatensystem** überträgt die Erdkugel auf eine Ebene, beispielsweise nach einer Kartenprojektion.

**Längengrad** und **Breitengrad** geben die Position auf einer (idealisierten) Kugel an. Der Breitengrad beschreibt dabei die Nord-Süd-Position, während der Längengrad die Ost-West-Position angibt. New York liegt etwa bei $40,7°$N und $73,9°$W und Ilmenau bei $50,7°$N und $10,9°$E. Um mathematisch positiv zu bleiben, werden Angaben in Richtung Westen auch häufig als negative Angaben in Richtung Osten notiert. So werden $73,9°$W zu $-73,9°$E.

Interessante **geometrische Objekte**, die auf Koordinaten basieren, sind
- **Punkte** *(Point)* - einzelne Orte (z.B. ein Taxi-Pickup),
- **Routen** *(Lines)* - eine Linie mit mehreren Punkten und
- **Polygone** - Flächen mit geschlossener Kontur (z.B. ein See).

## Historische Notizen zum Koordinatensystem
Eine der [ersten Bestimmungen des Breitengrads wird Eratosthenes](https://de.wikipedia.org/wiki/Eratosthenes#Bestimmung_des_Erdumfangs) (ca. 276 bis 194 v.Chr.) im antiken Griechenland zugeordnet. Er verwendete den Winkel der Sonnenhöhe in den Städten Alexandria und Syene und deren Abstand zueinander, um den Erdumfang zu berechnen. Der Längengrad wiederum war für Eratosthenes nicht berechenbar: Die fehlerhafte Annahme, die beiden Städte lägen auf dem selben Längengrad, gilt sogar als der schwerwiegendste Einfluss für die Abweichung seines Ergebnisses.

Claudius Ptolemäus [beschrieb im zweiten Jahrhundert n. Chr. zuerst den Quadranten](https://de.wikipedia.org/wiki/Quadrant_(Astronomie)#Geschichte). Mit diesem Instrument ließ sich mit Hilfe des Sonnenstands zur Mittagszeit oder des Polarsterns der Breitengrad bestimmen. Mit [einigen Verfeinerungen verbunden](https://de.wikipedia.org/wiki/Sextant) sollte diese Methode bis tief in die Neuzeit hinein sowohl an Land wie auf See zur Bestimmung des Breitengrads genügen, während die Bestimmung des Längengrads weiterhin ungelöst blieb.

Erste Ideen basierten auf Tabellen, die [verschiedene Winkel bezüglich des Mondes und Fixsternen ins Verhältnis setzten](https://de.wikipedia.org/wiki/Monddistanz#Verfahren_zur_L%C3%A4ngenbestimmung), um sowohl die Bestimmung des Längengrads wie auch der aktuellen (lokalen) Uhrzeit zu ermöglichen. Erst durch [John Harrisons Chronometer](https://de.wikipedia.org/wiki/John_Harrison_(Uhrmacher)#Harrisons_Vision) - informell formuliert eine auf den Bedarf der Royal Navy angepassten, recht genauen Uhr - ließ sich im 18. Jahrhundert die Zeit des Nullmeridians in Greenwich auf Reisen *mitnehmen*. Bei bekannter Ortszeit, beispielsweise durch Beobachtung des Sonnenstands, war die Ermittlung des Längengrads durch die Abweichung gegenüber dem Vereinigten Königreich endlich ein Leichtes.

Heute existiert mit [GPS](https://de.wikipedia.org/wiki/Global_Positioning_System) und vergleichbaren Systemen wie GLONASS, Galileo oder Beidou eine Möglichkeit zur zentimetergenauen, satellitengestützten Positionsbestimmung. Als Grundlage für die Form der Erde wird dabei ein sogenannter [Referenzellipsoid](https://de.wikipedia.org/wiki/World_Geodetic_System_1984) verwendet - auch vom Modell der Kugel hat die Menschheit also Abstand genommen.

## Entfernungsberechnung
Zur Berechnung der Entfernung auf einer Kugeloberfläche $d$ kann **nicht** der euklidische Abstand verwendet werden. Stellen Sie sich beispielsweise zwei Punkte vor, die genau auf gegenüberliegenden Seiten liegen: Der euklidische Abstand entspräche dann dem Durchmesser der Kugel.

Stattdessen existiert für diesen Anwendungsfall die sogenannte Haversine-Formel:

$$
d = 2r \; arcsin\left( \sqrt{sin^2\left( \frac{\left| \phi_1 - \phi_2 \right|}{2} \right) + cos(\phi_1) * cos(\phi_2) * sin^2\left( \frac{\left| \lambda_1 - \lambda_2 \right|}{2} \right)} \right)
$$

Dabei sind $\phi_1$ und $\phi_2$ die Breitengrade, $\lambda_1$ und $\lambda_2$ die Längengrade und $r$ der Kugelradius ($\approx 6371 \, km$).
