## GeoPandas
GeoPandas ist eine Python-Bibliothek zur einfachen Handhabung und Analyse von geographischen Daten. Sie erweitert die Funktionalität von Pandas um geometrische Operationen, indem sie auf leistungsfähige Bibliotheken wie *Shapely* und *Pyproj* zurückgreift, um das Lesen, Schreiben und Visualisieren von verschiedenen Vektorformaten mit der gewohnten Pandas Syntax zu ermöglichen.

GeoPandas lässt sich über das Paket `geopandas` laden und wird analog zu Pandas mit `gpd` abgekürzt.

```python
import geopandas as gpd
```

Weiterhin kann sich der Import spezifischer Klassen der von GeoPandas verwendeter Bibliotheken lohnen.

```python
from shapely.geometry import LineString, Point
```

## GeoSeries und GeoDataFrame
Die zentralen Klassen in GeoPandas verhalten sich analog zu denen in Pandas. `GeoSeries` stellt eine Spalte dar, die Geodaten enthält.

```python
gpd.GeoSeries([Point(1, 1), Point(2, 2), Point(3, 3)])
```

Mit dem optionalen Parameter `crs` (*Coordinate Reference System*) kann außerdem ein Referenzsystem gewählt werden. Standardmäßig wird [`EPSG:4326`](https://epsg.io/4326) verwendet - auch bekannt als WGS 84 und das Referenzsystem des GPS.

In einem `GeoDataFrame` können *reguläre* Spalten und solche mit GeoDaten gemischt werden.

```python
geo_df = gpd.GeoDataFrame({
    'name': ['Name 1',    'Name 2'],
    'geom': [Point(1, 2), Point(2, 1)]
})
geo_df
```

Spalten lassen sich wie gewohnt anhand ihres Namens referenzieren.

```python
geo_df['geom']
```

## Dateien lesen und speichern
GeoPandas erlaubt das Lesen und Schreiben vieler Formate mit *pyogrio* oder alternativ *fiona*. Die Funktion `list_drivers` gibt eine Übersicht über kompatible Formate in der aktuell installierten Version. (Einige Formate können nur gelesen `r` werden. Die meisten lassen sich aber lesen und schreiben `rw`.)

```python
import pyogrio

pyogrio.list_drivers()
```

Mit der Funktion `to_file` lässt sich ein GeoDataFrame in eine Datei schreiben. Standardmäßig wird dazu das ShapeFile-Format (*ESRI Shapefile*) verwendet. Durch Angabe eines gültigen Treiber lassen sich aber natürlich auch andere Formate speichern.

```python
geo_df.to_file('geodataframe.geojson', driver='GeoJSON')
```

Mit der Methode `read_file` lassen sich selbst geschriebene Dateien oder solche aus anderen Quellen einlesen.

```python
geo_df2 = gpd.read_file('geodataframe.geojson')
geo_df2
```

## Darstellung
GeoPandas interagiert zur Darstellung mit Matplotlib. Koordinaten mit Längen- und Breitengrad können einfach über die Funktion `plot` visualisiert werden.

```python
landschaft_df = gpd.GeoDataFrame({
    'fluesse': gpd.GeoSeries([
        LineString([ (7.0, 51.00), (7.1, 51.05), (7.2, 51.07), (7.3, 51.10) ]),
        LineString([ (7.0, 51.15), (7.1, 51.12), (7.2, 51.15), (7.3, 51.14) ])
    ])
})

landschaft_df['fluesse'].plot()
pass
```

Wie gewohnt lassen sich viele weitere Optionen von Matplotlib - wie Subplots, Farbe und Transparenz, Achsen-Optionen, etc. - verwenden.

## Operationen auf GeoDataFrames
GeoPandas besitzt [eine ausführliche Dokumentation](https://geopandas.org/en/stable/getting_started.html), welche die mannigfaltigen Funktionen auf Basis der Geodaten aufzeigt. An dieser Stelle soll deshalb nur ein kleines Beispiel gezeigt werden, dass die syntaktische Verwandtschaft zu Pandas demonstriert.

Die Funktion `buffer` erweitert eine Linie zu einem Polygon, indem eine (Puffer-) Zone in einem festen Abstand um die Linie hinzugefügt wird. Die Funktion wird auf eine GeoSeries angewendet und lässt sich dann auch wie gewohnt einer neuen Spalte des GeoDataFrames zuweisen.

```python
# 0.01 Grad ~ 1 km
landschaft_df['puffer'] = landschaft_df['fluesse'].buffer(0.01)
landschaft_df
```

Die Flüsse im Beispiel lassen sich dann mit der Pufferzone in eine gemeinsame Visualisierung zeichnen.

```python
import matplotlib.pyplot as plt

_, ax = plt.subplots()

landschaft_df['puffer'].plot(ax=ax, label='Puffer',
                            color='red', alpha=0.2, edgecolor='red')
landschaft_df['fluesse'].plot(ax=ax, label='Fluss',
                             color='darkblue', linewidth=3)

ax.set_xlabel('Länge')
ax.set_ylabel('Breite')

plt.show()
```

GeoPandas erweitert Pandas um Spalten, die Geodaten abbilden, und zugehörige Methoden. Die Bibliothek baut auf weiteren leistungsfähigen Bibliotheken auf und bietet den Zugriff auf die Daten in der gewohnten Syntax.
