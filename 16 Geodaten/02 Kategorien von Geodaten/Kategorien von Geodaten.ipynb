# Kategorien von Geodaten
Geodaten können nach unterschiedlichen Kriterium kategorisiert werden.

## Inhaltsverzeichnis
- [Vektor- und Rasterdaten](#Vektor--und-Rasterdaten)
- [Georeferenz](#Georeferenz)

## Vektor- und Rasterdaten
Die erste Kategorie wird durch die Datenstruktur definiert.

**Vektordaten** beschreiben reale Objekte durch Punkte (z.B. Taxistände), Linien (z.B. Straßen) oder Polygone (z.B. Waldflächen). Sie werden zum Beispiel als Koordinatenpaar ausgedrückt und erlauben eine präzise geometrische Erfassung. **Rasterdaten** hingegen verwenden ein Gitter (Raster) aus Zellen. Jeder Zelle wird dann ein Wert zugeordnet. Rasterdaten entstehen beispielsweise häufig bei der Erfassung durch Satelliten - das Raster ergibt sich durch die Auflösung der Sensoren, die Werte sind dann Temperaturen, Höhenwerte oder ähnliche Werte.

## Georeferenz
Die zweite Kategorie ergibt sich durch den Bezug zu einem eindeutigen Koordinatensystem.

**Georeferenzierte Daten** besitzen einen expliziten Bezug zu einem definierten Koordinatensystem. Währenddessen haben **nicht-georeferenzierte Daten** zwar einen räumlichen Bezug, allerdings keine konkreten Koordinaten. Beispielsweise müssen Adressen erst durch eine Rückwärtssuche mit Koordinaten verknüpft werden, um diese raumbezogene Information tatsächlich nutzbar zu machen.
