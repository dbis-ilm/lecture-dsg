## Kommentare
Kommentare im Quellcode sind immens wichtig. Sie ermöglichen es, Gedanken und Anmerkungen, die nicht direkt ersichtlich sind, festzuhalten ohne das Programm zu verändern. Die meisten deutschen oder englischen Sätze stellen darüber hinaus keinen syntaktisch korrekten Quellcode dar, sodass gar keine Ausführung möglich wäre.

In Python existieren (wie in vielen Programmiersprachen) zwei verschiedene Arten von Kommentaren.

````python
# Einzeilige Kommentare werden mit einer Raute eingeleitet
# und schließen die aktuelle Zeile von der Ausführung aus.
a = 1

# Sie können auch verwendet werden, um einzelne Anweisungen
# "auszukommentieren". Nach dem Ausführen dieser Zelle werden
# Sie feststellen, dass `a` den Wert `1` besitzt.

# a = 2

a
````

````python
"""
Mehrzeilige Kommentare werden durch drei Anführungszeichen
eingeschlossen. Somit muss nicht jede Zeile einzeln mit
einer Raute versehen werden.
"""
b = 3

'''
Welche Anführungszeichen verwendet werden, ist nicht relevant.
Es müssen jedoch die selben Anführungszeichen zum Schließen
verwendet werden, die auch zum Öffnen des Kommentarbereichs
verwendet wurden.

b = 4
'''

b
````

## Einrückungen und Blöcke
Ein **Block** ist ein Abschnitt innerhalb eines Programms, der zusammengehörige Anweisungen bündelt, und damit die logische Struktur eines Programms bestimmt. In Python beginnen Blöcke in der Regel durch eine Anweisung, die mit einem Doppelpunkt beendet wird. Die darauffolgenden Anweisungen werden durch Einrückungen mit vier Leerzeichen oder einem Tabstopp gekennzeichnet. Sobald die Einrückung beendent wird, gilt auch der Block als beendet. Fehlt die Einrückung oder ist sie inkonsistent, führt das zu einem *SyntaxError*, da die Anweisungen nicht eindeutig zugeordnet werden können.

```python
Anweisung 1

Anweisungskopf:
    Anweisung 2.1
    Anweisung 2.2
    ...
    Anweisung 2.n

Anweisung 3
```

In vielen Sprachen werden zur Kennzeichnung geschweifte Klammern (`{}`) oder spezielle Schlüsselworte (`begin`, `end`) verwendet. Python dagegen erzwingt durch die Verwendung von Einrückungen auch gleichzeitig das Schreiben von lesbarerem Code, was besonders Programmieranfängern entgegenkommt.

Später werden Sie einige Kontrollstrukturen kennenlernen, die Blöcke verwenden und diese auch ineinander stapeln. Nachfolgend ist eine als Beispiel eine Zählschleife dargestellt, wobei der ihr zugehörige, eingerückte Code zehn Mal ausgeführt wird. Währenddessen wird der auf die Schleife nachfolgende, nicht eingerückte Code nur einmalig ausgeführt, da er nicht mehr zur Schleife zählt.

````python
for i in range(10):
    # Der nachfolgende Code ist eingerückt
    # und gehört damit zum Schleifenrumpf.
    # Demnach wird er zehn Mal ausgeführt.
    print(i)

# Die nachfolgende Zeile ist nicht mehr
# eingerückt und wird daher nur einmalig
# ausgeführt.
print(100)
````
